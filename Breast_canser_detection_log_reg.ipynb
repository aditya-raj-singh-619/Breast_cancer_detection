```python
# importing the librareis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
```

```python
# importing the dataset
dataset=pd.read_csv('breast_cancer.csv')
X=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

```python
# spiliting into trainig and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2,random_state=0)
print(X_train)
```

```python
#traininng the logistic regression model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
#predictting the test set result
y_pred=classifier.predict(X_test)
print(y_pred)
```

```python
#making the confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
#accuracy test
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```

```python
# k-fold validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
# load the model and it can be for new dataset
classifier = joblib.load('logistic_model.pkl') 
new_dataset = pd.read_csv('new_datasets.csv')  
X_new = new_dataset.values 

y_pred = classifier.predict(X_new)
print("Predicted Output:\n", y_pred)

df_with_predictions = new_dataset.copy()
df_with_predictions['Predicted_Target'] = y_pred 

# Save to CSV
df_with_predictions.to_csv('new_dataset_with_predictions.csv', index=False)

print("Predictions saved to 'new_dataset_with_predictions.csv'")

```
